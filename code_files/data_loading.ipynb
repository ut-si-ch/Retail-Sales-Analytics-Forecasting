## Insights from Data Inspection

#### 1) Features DataSet(8190 rows, 12 columns)
   - Markdown 1-5 -> ~50% missing values(expected that campaigns don't run every week)
   - CPI & Unemployment -> 585 missing values(~7%)
   - 'Date' column need to be converted from object to date format.
     
   - **Temperature**
        - Avg ~59°F, range –7°F (very cold) to 102°F (very hot).
        - Stores operate in very diverse climates → sales may depend on weather (e.g., hot weather = higher beverage/AC sales).

   - **Fuel Price**
        - Range: $2.47 – $4.47.
        - Fuel cost fluctuations may impact customer spending power → sales might dip when fuel is expensive.

   - **MarkDown1–5 (promotional discounts)**
        - Extremely skewed with huge outliers (Max up to 771,448) → only active in some weeks, absent in many (missing values).
        - Promotions likely drive sales spikes → important predictor.

   - **CPI (Consumer Price Index)**
        - Range: 126 – 229, std ~39 → stores span different economic environments.
        - Higher CPI → higher living costs → may reduce discretionary purchases.

   - **Unemployment**
        - Range: 3.68% – 14.31%, mean ~7.8%.
        - Big spread → local economy strongly influences store sales.

   - **IsHoliday**
        - 7605 False vs 585 True → ~7% are holidays.
        - Sales spikes should align with these holiday weeks.


#### 2) **Sales Dataset(421570 rows, 5 columns)**

   - No missing values.
   - Columns - Store, Dept, Date, Weekly_Sales, IsHoliday.
   - Sales are by Store + Dept + Date
   - Date column aligns with features dataset so will be key for merging.
   - **Weekly Sales**
     - Mean = $15.9k, but max = $693k, min = –$4.9k (returns/refunds).
     - Very high variance → model must handle skewness.
     - Median ~$7.6k → sales distribution is right-skewed (few very high values).

   - **Seasonality**
     - Top date is 23-Dec-2011, right before Christmas → confirms strong holiday seasonality.
     - Holiday effects are crucial features.

   - **Store / Dept**
     - Stores 1–45, Depts 1–99 → not all departments are equally profitable.
     - Some departments likely dominate (e.g., groceries).
   
#### 4) **Stores Dataset(45 rows, 3 columns)**

   - No missing values.
   - Columns: Store, Type(A/B/C), Size.
   - **Store Size**
     - Avg ~130k sq.ft., range 34k – 219k.
     - Larger stores (likely “Type A”) may generate more sales.
   - **Type**
     - Three categories: A, B, C.
     - A most common (22 stores).
     - Likely: Type A = supercenters, Type B = medium, Type C = smaller.
